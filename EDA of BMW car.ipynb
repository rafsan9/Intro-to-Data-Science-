```python
df['transmission'].value_counts().plot.pie(autopct = '%.2f%%')
```

```python
df['transmission'].value_counts().plot(kind = 'pie',
                                     autopct = '%.1f%%')
plt.show()
```

```python
plt.figure(figsize=(20,16))

plt.subplot(3, 1, 1)
sns.scatterplot(x="mileage", y="price", data=df)

plt.subplot(3, 1, 2)
sns.scatterplot(x="mpg", y="price", data=df)

plt.subplot(3, 1, 3)
sns.scatterplot(x="engineSize", y="price", data=df)

plt.tight_layout()
plt.show()
```

```python
# Create a regression plot
sns.regplot(x='price', y='mileage', data=df)

# Display the plot
sns.plt.show()
```

```python
# Create a regression plot
sns.regplot(x='price', y='engineSize', data=df)

# Display the plot
sns.plt.show()
```

```python
sns.pairplot(df)
```

```python
plt.figure(figsize=(20,16))

plt.subplot(2, 2, 1)
sns.scatterplot(x="mileage", y="price", data=df, hue="fuelType")

plt.subplot(2, 2, 2)
sns.scatterplot(x="mpg", y="price", data=df, hue="fuelType")

plt.subplot(2, 2, 3)
sns.scatterplot(x="engineSize", y="price", data=df, hue="fuelType")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(y = "price", data=df, showfliers=False)
```

```python
plt.figure(figsize=(20,8)) # width, height
sns.violinplot(x='price', data=df, showfliers=False)

plt.suptitle("Violinplot of the price columns ")
plt.tight_layout()
plt.show()
```

```python
raf=df.groupby('transmission')[['mileage','price']].sum().sort_values(by='price',ascending=False)
raf[:].plot.bar(color=['red','blue'],figsize=(8,6))
plt.title('Price increase or decrease with mileage')
plt.xlabel('transmission')
plt.ylabel('price')
plt.show()
```

```python
plt.figure(figsize=(12,10))
plt.title('Price in year')
sns.barplot(x='year',y='price',data=df,order=df['year'].value_counts().index,palette='rocket')
plt.xlabel('year',fontsize=10)
plt.show()
```

```python
raf=df.groupby('model')[['price']].sum().sort_values(by='price',ascending=False)
raf[:].plot.bar(color=['red','blue'],figsize=(8,6))
plt.title('Price increase or decrease with model')
plt.xlabel('model')
plt.ylabel('price')
plt.show()
```
